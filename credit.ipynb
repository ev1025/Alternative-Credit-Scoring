```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

def create_core_features(data, mapping):
    core_features = {}
    scaler = MinMaxScaler()
    
    for feature_name, columns in mapping.items():
        
        # 해당 그룹의 컬럼들 정규화
        scaled_values = scaler.fit_transform(data[columns])
        # 정규화된 값들의 평균 계산
        core_features[feature_name] = np.mean(scaled_values, axis=1)
    
    return pd.DataFrame(core_features)

# 인덱스 확인 및 리셋
credit_index = create_core_features(user_card_df, credit_mapping).reset_index(drop=True)

# cleaned 데이터도 인덱스 리셋
predict_data = user_card_df.reset_index(drop=True)

# 결합 시도
credit_raw = pd.concat([credit_index, user_card_df[['남성', '여성', '20대', '30대', '40대', '50대', '60대', '70대', '회원여부_연체']]], axis=1)

# 결과 확인
display(credit_raw.head())
```
