```python
## _을 NaN으로 변경
csv_file_copy_df.replace('_', np.nan, inplace=True)
csv_file_copy_df
```

## 남녀구분코드

```python
## 1 --- 남성, 2 --- 여성 으로 변경
csv_file_copy_df['남녀구분코드'] = np.where(csv_file_copy_df['남녀구분코드'] == 1, '남성', '여성')
# csv_file_copy_df['남녀구분코드'] = np.where(csv_file_copy_df['남녀구분코드'] == '남성', 1, 2)
csv_file_copy_df
```

```python
gender = pd.get_dummies(csv_file_copy_df['남녀구분코드'])
csv_file_copy_df.drop(['남녀구분코드'], axis=1, inplace=True)
csv_file_copy_df.rename(columns={'남성': 1, '여성': 2}, inplace=True)
csv_file_copy_df = pd.concat((csv_file_copy_df, gender), axis=1)
csv_file_copy_df
```

## 가입통신회사코드

```python
# KT - 0 / SK - 1 / LG - 2 / 기타 - 3

csv_file_copy_df['가입통신회사코드'].value_counts()
csv_file_copy_df['가입통신회사코드'].replace('KTF', 'KT', inplace=True)
csv_file_copy_df['가입통신회사코드'].replace('SKT', 'SK', inplace=True)
csv_file_copy_df['가입통신회사코드'].replace('LGT', 'LGU+', inplace=True)
csv_file_copy_df['가입통신회사코드'].replace('ETC', '기타', inplace=True)
csv_file_copy_df['가입통신회사코드']
```

```python
bs_name = pd.get_dummies(csv_file_copy_df['가입통신회사코드'])
csv_file_copy_df.drop('가입통신회사코드', axis=1, inplace=True)
csv_file_copy_df = pd.concat([csv_file_copy_df, bs_name], axis=1)
csv_file_copy_df
```

```python
card_name = pd.get_dummies(csv_file_copy_df['_1순위신용체크구분'])
csv_file_copy_df.drop('_1순위신용체크구분', axis=1, inplace=True)
csv_file_copy_df = pd.concat([csv_file_copy_df, card_name], axis=1)
csv_file_copy_df
```
